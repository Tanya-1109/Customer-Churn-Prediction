```python
import sys
import os
sys.path.append(os.path.abspath(".."))  # Links the parent directory to the notebook's environment

import pandas as pd
import joblib

pipeline = joblib.load("../models/churn_pipeline.pkl")

# Simulate a raw customer record exactly as it comes from the source system
sample_customer = pd.DataFrame([{
    'customerID': 'TEST-001',
    'gender': 'Female',
    'SeniorCitizen': 0,
    'Partner': 'Yes',
    'Dependents': 'No',
    'tenure': 3,
    'PhoneService': 'Yes',
    'MultipleLines': 'No',
    'InternetService': 'Fiber optic',
    'OnlineSecurity': 'No',
    'OnlineBackup': 'No',
    'DeviceProtection': 'No',
    'TechSupport': 'No',
    'StreamingTV': 'Yes',
    'StreamingMovies': 'Yes',
    'Contract': 'Month-to-month',
    'PaperlessBilling': 'Yes',
    'PaymentMethod': 'Electronic check',
    'MonthlyCharges': 85.0,
    'TotalCharges': '255.0'
}])


churn_proba = pipeline.predict_proba(sample_customer)[:, 1][0]
print(f"Churn probability: {churn_proba:.2%}")
print(f"Prediction: {'HIGH RISK' if churn_proba > 0.35 else 'LOW RISK'}")
```
